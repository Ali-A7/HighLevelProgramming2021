```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


mean1=0
dev1=0.4
no1=200
mean2=1
dev2=0.2
no2=100
a1 = np.random.normal(loc=mean1, scale=dev1, size=(2*n1,))
a2 = np.random.normal(loc=mean1, scale=dev2, size=(2*n2,))
val1=np.reshape(a1,(no1,2))
val2=np.reshape(a2,(no2,2))
fig, ax = plt.subplots()
ax.scatter(x=val1[:,0], y=val1[:,1], c="indigo",alpha=0.85, s=99, label="Gaus(0,0.5)")
ax.scatter(x=val2[:,0], y=val2[:,1], c="gold",alpha=0.85, s=99, label="Gaus(1,0.2)")
plt.legend(loc="upper left")
plt.show()
```
